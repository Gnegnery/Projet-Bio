## Analise général des séquences

```python
print("Nombre séquences:", len(peaks_seq))

ratio_nuc = ms.getRatio(peaks_seq)
nucs = "ATCG"
nuc_tot = ratio_nuc[2]
print("Ratio nucléotides:")
for i in range(4):
    nuc = nucs[i]
    nuc_nb = ratio_nuc[0][nuc]
    nuc_per = ratio_nuc[1][i]
    print(f"{nuc}: {nuc_nb} {nuc_per:f}%")

print("totale:", nuc_tot, "nucléotides")
```
